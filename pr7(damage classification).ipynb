```python
import numpy as np

# Example Structural Data
# Feature1 = vibration
# Feature2 = strain
# 0 = No Damage
# 1 = Damage

X = np.array([
    [0.1, 0.2],
    [0.15, 0.25],
    [0.2, 0.3],
    [0.8, 0.9],
    [0.75, 0.85],
    [0.9, 0.95]
])

y = np.array([0, 0, 0, 1, 1, 1])

# Step 1: Generate Immune Detectors
num_detectors = 5
detectors = np.random.rand(num_detectors, 2)

print("Generated Detectors:\n", detectors)

# Step 2: Distance Function
def distance(a, b):
    return np.sqrt(np.sum((a - b) ** 2))

# Step 3: Prediction Function
def predict(sample):
    threshold = 0.5
    for d in detectors:
        if distance(sample, d) < threshold:
            return 1  # Damage detected
    return 0  # No damage

# Step 4: Test the Model
print("\nClassification Results:")
for i in range(len(X)):
    pred = predict(X[i])
    print("Input:", X[i], "Predicted:", pred, "Actual:", y[i])
```
