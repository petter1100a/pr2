```python
import random
import time

# ----- SERVER CLASS -----
class Server:
    def __init__(self, server_id):
        self.server_id = server_id
        self.active_requests = 0

    def process_request(self):
        self.active_requests += 1

    def finish_request(self):
        if self.active_requests > 0:
            self.active_requests -= 1


# ----- LOAD BALANCER -----
class LoadBalancer:
    def __init__(self, servers):
        self.servers = servers
        self.rr_index = 0

    # Round Robin
    def round_robin(self):
        server = self.servers[self.rr_index]
        self.rr_index = (self.rr_index + 1) % len(self.servers)
        return server

    # Random
    def random_choice(self):
        return random.choice(self.servers)

    # Least Connections
    def least_connections(self):
        return min(self.servers, key=lambda s: s.active_requests)


# ----- SIMULATION -----
def simulate_requests(num_requests, algorithm):
    servers = [Server(i) for i in range(1, 4)]
    lb = LoadBalancer(servers)

    print(f"\nLoad Balancing Algorithm: {algorithm}")
    print("-" * 40)

    for req in range(1, num_requests + 1):
        if algorithm == "round_robin":
            server = lb.round_robin()
        elif algorithm == "random":
            server = lb.random_choice()
        elif algorithm == "least_connections":
            server = lb.least_connections()
        else:
            return

        server.process_request()
        print(f"Request {req} -> Server {server.server_id}")

        # Simulate completion
        if random.choice([True, False]):
            server.finish_request()

        time.sleep(0.2)

    print("\nFinal Server Loads:")
    for s in servers:
        print(f"Server {s.server_id}: {s.active_requests} active requests")


# ----- MAIN -----
if __name__ == "__main__":
    print("LOAD BALANCING SIMULATION")
    print("1. Round Robin")
    print("2. Random")
    print("3. Least Connections")

    choice = input("\nEnter your choice (1/2/3): ")

    if choice == "1":
        simulate_requests(10, "round_robin")
    elif choice == "2":
        simulate_requests(10, "random")
    elif choice == "3":
        simulate_requests(10, "least_connections")
    else:
        print("Invalid choice!")
```
