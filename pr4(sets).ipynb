```python
# ----- Fuzzy Set Operations -----

def fuzzy_union(A, B):
    return {x: max(A.get(x, 0), B.get(x, 0)) for x in set(A) | set(B)}

def fuzzy_intersection(A, B):
    return {x: min(A.get(x, 0), B.get(x, 0)) for x in set(A) | set(B)}

def fuzzy_complement(A):
    return {x: round(1 - A[x], 2) for x in A}

def fuzzy_difference(A, B):
    return {x: min(A.get(x, 0), 1 - B.get(x, 0)) for x in A}

# ----- Cartesian Product -----

def cartesian_product(A, B):
    relation = {}
    for x in A:
        for y in B:
            relation[(x, y)] = min(A[x], B[y])
    return relation

# ----- Max-Min Composition -----

def max_min_composition(R1, R2, X, Y, Z):
    R = {}
    for x in X:
        for z in Z:
            min_values = []
            for y in Y:
                min_val = min(R1.get((x, y), 0), R2.get((y, z), 0))
                min_values.append(min_val)
            R[(x, z)] = max(min_values)
    return R

# ----- Display Functions -----

def display_set(title, A):
    print("\n", title)
    for k, v in A.items():
        print(f"{k} : {v}")

def display_relation(title, R):
    print("\n", title)
    for k, v in R.items():
        print(f"{k} : {v}")

# ----- Main Program -----

def main():
    A = {'x1': 0.2, 'x2': 0.5, 'x3': 0.8}
    B = {'x1': 0.6, 'x2': 0.4, 'x3': 0.3}

    print("Fuzzy Set A:", A)
    print("Fuzzy Set B:", B)

    union = fuzzy_union(A, B)
    intersection = fuzzy_intersection(A, B)
    complementA = fuzzy_complement(A)
    difference = fuzzy_difference(A, B)

    display_set("Union (A ∪ B)", union)
    display_set("Intersection (A ∩ B)", intersection)
    display_set("Complement of A", complementA)
    display_set("Difference (A - B)", difference)

    # Relations
    C = {'y1': 0.3, 'y2': 0.7}
    D = {'z1': 0.4, 'z2': 0.9}

    print("\nFuzzy Set C:", C)
    print("Fuzzy Set D:", D)

    R1 = cartesian_product(A, C)
    R2 = cartesian_product(C, D)

    display_relation("Relation R1 (A x C)", R1)
    display_relation("Relation R2 (C x D)", R2)

    R = max_min_composition(R1, R2, A, C, D)
    display_relation("Max-Min Composition", R)

# FIXED MAIN
if __name__ == "__main__":
    main()
```
