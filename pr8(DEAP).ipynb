```python
from deap import base, creator, tools, algorithms
import random

# ----- Step 1: Define Fitness & Individual -----
creator.create("FitnessMax", base.Fitness, weights=(1.0,))
creator.create("Individual", list, fitness=creator.FitnessMax)

# ----- Step 2: Toolbox -----
toolbox = base.Toolbox()
toolbox.register("attr_bool", random.randint, 0, 1)

toolbox.register("individual", tools.initRepeat, creator.Individual, toolbox.attr_bool, 20)
toolbox.register("population", tools.initRepeat, list, toolbox.individual)

# ----- Step 3: Fitness Function -----
def evalOneMax(individual):
    return (sum(individual),)

toolbox.register("evaluate", evalOneMax)

# ----- Step 4: Operators -----
toolbox.register("mate", tools.cxTwoPoint)
toolbox.register("mutate", tools.mutFlipBit, indpb=0.1)
toolbox.register("select", tools.selTournament, tournsize=3)

# ----- Step 5: Main -----
def main():
    pop = toolbox.population(n=50)

    stats = tools.Statistics(lambda ind: ind.fitness.values)
    stats.register("avg", lambda x: sum(i[0] for i in x) / len(x))
    stats.register("max", lambda x: max(i[0] for i in x))

    pop, log = algorithms.eaSimple(
        pop,
        toolbox,
        cxpb=0.5,
        mutpb=0.2,
        ngen=10,
        stats=stats,
        verbose=True
    )

    best = tools.selBest(pop, k=1)[0]
    print("\nBest Individual:", best)
    print("Fitness:", best.fitness.values[0])

if __name__ == "__main__":
    main()
```
